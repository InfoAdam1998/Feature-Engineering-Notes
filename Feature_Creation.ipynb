# Compare features to reference variable


```python
#Step 1: Import RelativeFeatures from feature-engine
# This will allow the creation of new features by calculating relative relationships (e.g., subtraction or division) between variables
from feature_engine.creation import RelativeFeatures

#Step 2: Define the features of interest
# These are the variables that will be compared against the reference variables
features = ["mean smoothness", "mean compactness", "mean concavity", "mean symmetry"]

#Step 3: Define the reference features
# These are the reference variables used in the relative calculations with the features of interest
reference = ["mean radius", "mean area"]

#Step 4: Initialize the RelativeFeatures transformer
# Apply subtraction and division between the features of interest and the reference features
creator = RelativeFeatures(
    variables=features,            # Specify features to compare
    reference=reference,           # Define reference features to compare against
    func=["sub", "div"],           # Functions to apply: subtraction and division
)

#Step 5: Fit and transform the dataset
# Apply the relative transformations on the dataset and create new features
df_t = creator.fit_transform(df)

#Step 6: Get the names of the new features
# Extract names of the newly created features by filtering out the original feature names
new_features = [f for f in df_t.columns if f not in creator.feature_names_in_]

#Step 7: Display the new relative features
# Show the first few rows of the newly created features
df_t[new_features].head()
```
